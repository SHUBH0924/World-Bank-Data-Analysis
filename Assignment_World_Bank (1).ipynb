# Data Preprocessing

```python
metadata_country = metadata_country[['Country Code','Region']]
```

**To automate the process we develop function**

```python
def data_processing(df, value_name):
  years = [str(i) for i in range(1960, 2017)]
  df.drop(['Country Name', 'Indicator Name', 'Indicator Code'], axis=1, inplace=True)
  df.dropna(axis=0,inplace=True)
  df = df.melt(id_vars='Country Code', value_vars=years, var_name='Year', value_name=value_name)
  return df
```

```python
country_population = data_processing(country_population, 'Population')
fertility_rate = data_processing(fertility_rate, 'Fertility Rate')
life_expectancy = data_processing(life_expectancy, 'Life Expectancy')
```

# Merge

```python
df = pd.merge(metadata_country, country_population, on='Country Code', how='left')
df.isna().sum()
```

```python
df = pd.merge(df, life_expectancy, how='left', on=['Country Code', 'Year'])
df.isna().sum()
```

```python
df = pd.merge(df, fertility_rate, how='left', on=['Country Code', 'Year'])
df.isna().sum()
```

```python
df.shape
```

```python
df.isna().sum().sum()
```

```python
df['Population'] = df['Population']/1000
df.head()
df.tail()
```

# Handling Missing Values

```python
df.dropna(axis=0, inplace=True)
df.isna().sum().sum()
```

```python
df.shape
```

# Visualization

```python
fig = px.line(df, x='Region',
              y='Population',
              color='Region',
              animation_frame="Year",
              animation_group="Country Code",
              title='Population growth by Region',
              range_y = [0,1500])
fig.show()
```

```python
fig = px.histogram(df,x='Region',
                   y='Fertility Rate',
                   color='Region',
                   title='Fertility rate by Region',
                   animation_frame='Year',
                   animation_group='Region',
                   range_y=[0,400]
                   )
fig.show()
```

```python
px.scatter(df,
           x="Fertility Rate",
           y="Life Expectancy",
           animation_frame="Year",
           animation_group="Country Code",
           size="Population",
           color="Region",
           hover_name="Country Code",
           log_x=True,
           size_max=55,
           range_x=[1,10],
           range_y=[10,100])
```

```python
fig = px.box(df,x='Region',
             y='Life Expectancy',
             color='Region',
             title='Life Expectancy by Region',
             animation_frame='Year',
             animation_group='Country Code',
             range_y = [0,90]
)
fig.show()
```

```python
grouped_data = df.groupby(["Region", "Year"])[["Population", "Life Expectancy", "Fertility Rate"]].mean().reset_index()

fig = px.line(grouped_data,
              x="Year",
              y="Life Expectancy",
              color="Region",
              title="Life Expectancy by Region and Year",
              range_y=[0,90],)
fig.show()
```
